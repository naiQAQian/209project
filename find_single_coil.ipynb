```python
import h5py
import os
import shutil 
#hi
```

```python
# Function to check if the file is multi-coil or uses parallel imaging
def is_parallel_or_multicoil(hf):
    # Check the k-space shape
    kspace = hf['kspace'][()]
    kspace_shape = kspace.shape

    # Check if the k-space has a coil dimension (e.g., (num_coils, ...))
    if len(kspace_shape) >= 4:  # Assuming multi-coil data has at least 4 dimensions
        print(f"File is multi-coil with k-space shape: {kspace_shape}")
        return True

    # If neither multi-coil nor parallel imaging is detected
    return False
```

```python
# Define the source directory and destination directory
source_directory = r'D:\Class Project\209\brain_test_batch_0'
destination_directory = r'D:\Class Project\209\non_parallel'

for filename in os.listdir(source_directory):
    if filename.endswith(".h5"):
        file_path = os.path.join(source_directory, filename)
        
        # Open the HDF5 file and process it
        with h5py.File(file_path, 'r') as hf:
            print(f"Processing file: {filename}")

            # Check if the file is multi-coil or parallel imaging
            if not is_parallel_or_multicoil(hf):
                # Move the file to the new folder if it's not multi-coil or parallel imaging
                destination_path = os.path.join(destination_directory, filename)
                print(f"Moving {filename} to {destination_directory}")
                shutil.move(file_path, destination_path)
            else:
                print(f"{filename} contains multi-coil or parallel imaging data.")

        # The file will automatically close after the `with` block
        print(f"Finished processing {filename}\n")
```
